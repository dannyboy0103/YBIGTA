```python
plt.figure(figsize = (10, 10))
for i, col in enumerate(data.columns.to_list()):
  plt.subplot(5, 5, i + 1)
  plt.hist(data[col], label = col, color = 'black')
  plt.title(col)
  plt.tight_layout()
```

```python
plt.figure(figsize = (20, 20))
sns.heatmap(data.corr(), annot=True, cmap='Greens')
```
